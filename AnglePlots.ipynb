```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from glob import glob
#plt.rcParams["figure.figsize"] = (15,10)
```

```python
src       = "/run/media/maximilian/2E77B850185279B3/MC_DATA/data_normal/sorted/small_runs"
#src_hel   = src + "/Helix_theta/out*"
#src_rec   = src + "/Rec_theta/out*"
#src_truth = src + "/Truth_theta/out*"

src_hel   = src + "/Helix_phi/out*"
src_rec   = src + "/Rec_phi/out*"
src_truth = src + "/Truth_phi/out*"

fnames_angle_hel   = glob(src_hel)
fnames_angle_rec   = glob(src_rec)
fnames_angle_truth = glob(src_truth)

arrays_hel   = [np.load(f) for f in fnames_angle_hel]
arrays_rec   = [np.load(f) for f in fnames_angle_rec]
arrays_truth = [np.load(f) for f in fnames_angle_truth]
```

```python
binned_hel = 0
xedges_hel = arrays_hel[0]["xedges"]
yedges_hel = arrays_hel[0]["yedges"]
for i in range(6):
    binned_hel += arrays_hel[i]["data"]

binned_rec = 0
xedges_rec = arrays_rec[0]["xedges"]
yedges_rec = arrays_rec[0]["yedges"]
for i in range(6):
    binned_rec += arrays_rec[i]["data"]

binned_truth = 0
xedges_truth = arrays_truth[0]["xedges"]
yedges_truth = arrays_truth[0]["yedges"]
for i in range(6):
    binned_truth += arrays_truth[i]["data"]
```

```python
binned = binned_truth# - binned_hel
xedges = xedges_truth
yedges = yedges_truth
```

```python
fig = plt.figure(figsize=(10, 8))

ax = fig.add_subplot(111)
X, Y = np.meshgrid(xedges, yedges)
im = ax.pcolormesh(X, Y, binned.T, cmap="turbo", vmin=0, vmax=120)
plt.title("phi distribution (mc truth)", size=20)
plt.ylabel("phi", size=20)
plt.xlabel("z", size=20)
plt.colorbar(im)
plt.tight_layout()
plt.savefig("angle_phi_dist_from_truth", dpi=300)
```
